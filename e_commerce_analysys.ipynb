```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
customers = pd.read_csv('olist_customers_dataset.csv', sep=',')

order_items = pd.read_csv('olist_order_items_dataset.csv', sep=',')

orders = pd.read_csv('olist_orders_dataset.csv', sep=',')
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp']).dt.date
```

```python
user_orders = orders[['customer_id', 'order_id']].copy().merge(customers[['customer_id', 'customer_unique_id']].copy(), on='customer_id')
single_purchase_users = user_orders.groupby('customer_unique_id')['order_id'].nunique()
single_purchase_count = single_purchase_users[single_purchase_users == 1].count()
print(f"Количество пользователей, совершивших только одну покупку: {single_purchase_count}")
daily_orders = orders.groupby('order_purchase_timestamp').size().reset_index(name='order_count')
daily_orders['order_purchase_timestamp'] = pd.to_datetime(daily_orders['order_purchase_timestamp'])

plt.figure(figsize=(14, 6))
sns.lineplot(data=daily_orders, x='order_purchase_timestamp', y='order_count')
plt.title('Количество заказов по дням')
plt.xlabel('Дата')
plt.ylabel('Количество заказов')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
failed_deliveries = orders.copy()
failed_deliveries['order_purchase_month'] = pd.to_datetime(failed_deliveries['order_purchase_timestamp']).dt.to_period('M')
failed_deliveries = failed_deliveries[failed_deliveries['order_status'].isin(['canceled', 'unavailable'])]
failed_deliveries_count = failed_deliveries.groupby(['order_purchase_month', 'order_status']).size().unstack(fill_value=0)
monthly_avg_failed_deliveries = failed_deliveries_count.mean()
print("Среднее количество недоставленных заказов по месяцам с детализацией причин:")
print(monthly_avg_failed_deliveries)
failed_deliveries_count = failed_deliveries_count.reset_index()

# Преобразование столбца 'order_purchase_month' в формат datetime для правильного отображения на графике
failed_deliveries_count['order_purchase_month'] = failed_deliveries_count['order_purchase_month'].astype(str).apply(lambda x: pd.to_datetime(x))

# Построение графика линий для отображения количества недоставленных заказов по месяцам
plt.figure(figsize=(14, 6))
sns.lineplot(data=failed_deliveries_count.set_index('order_purchase_month'), markers=True)
plt.title('Количество недоставленных заказов по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Количество заказов')
plt.xticks(rotation=45)
plt.legend(title='Статус заказа')
plt.tight_layout()
plt.show()
```

```python
items_orders = order_items.copy().merge(orders[['order_id', 'order_purchase_timestamp']].copy(), on='order_id')
items_orders['day_of_week'] = pd.to_datetime(items_orders['order_purchase_timestamp']).dt.day_name()
most_frequent_day = items_orders.groupby('product_id')['day_of_week'].agg(lambda x: x.mode()[0])
print("День недели с наибольшей частотой покупок для каждого товара:")
print(most_frequent_day)
day_frequency = most_frequent_day.value_counts()

# Устанавливаем порядок дней недели
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Сортируем данные по порядку дней недели
day_frequency = day_frequency.reindex(days_order)

# Построение гистограммы распределения количества товаров по дням недели
plt.figure(figsize=(10, 6))
sns.barplot(x=day_frequency.index, y=day_frequency.values)
plt.title('Распределение количества товаров по дням недели с наибольшей частотой покупок')
plt.xlabel('День недели')
plt.ylabel('Количество товаров')
plt.xticks(rotation=45)
plt.show()
```
